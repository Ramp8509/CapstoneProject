# Step 1: Problem Understanding & Framing

**Business Problem :**

Financial institutions process millions of transactions every day. Even though fraudulent transactions represent a tiny fraction of all transactions, they result in significant financial losses and damage customer trust.

The goal of this project is to build a machine learning model that can accurately identify fraudulent credit card transactions while minimizing false positives, ensuring legitimate customer transactions are not unnecessarily blocked.



**Data Science :**

This is a binary classification problem.

Target Variable:

Class

0 = Legitimate Transaction

1 = Fraudulent Transaction

**Classification**:


This project aims to classify financial transactions as either legitimate or fraudulent based on historical transaction data. Since the dataset includes labeled examples (Class = 0 for legitimate and Class = 1 for fraudulent transactions), the problem is formulated as a supervised binary classification task. Although fraud detection is closely related to anomaly detection because fraudulent transactions are rare and unusual, the availability of labeled data makes supervised classification the most appropriate approach for this project.

Note: While this project is implemented as a supervised binary classification task, credit card fraud detection is also commonly approached as an anomaly detection problem in real-world scenarios where labeled fraud data is limited or unavailable. Future work could compare supervised classification models with anomaly detection techniques such as Isolation Forest or One-Class SVM.
